```python
def createKeypointArrays(keypointArray):
    xArray = []
    yArray = []
    confidences = []
    for i in range (len(keypointArray)):
        if(i%3 == 0):
            xArray.append(keypointArray[i])
        elif(i%3 == 1):
            yArray.append(keypointArray[i])
        else:
            confidences.append(keypointArray[i])
    
    figure1 = plt.subplot(131)
    figure1.plot(xArray,yArray, 'o', c= 'red')
    
    figure1.set_title("Raw keypoint positions")
    figure1.set_ylim([1,0])
    figure1.set_xlim([0,1])
    figure1.set_aspect('equal')
    
    xmin = min(xArray)
    xmax = max(xArray)
    ymin = min(yArray)
    ymax = max(yArray)
    for i in range(len(xArray)):
        xArray[i] = (xArray[i] - xmin)/(xmax-xmin)
        yArray[i] = (yArray[i] - ymin)/(ymax-ymin)
    
    figure2 = plt.subplot(132)
    figure2.plot(xArray,yArray, 'o', c= 'blue')
    figure2.set_title("Bounded and scaled keypoint positions")
    figure2.set_ylim([1,0])
    figure2.set_xlim([0,1])
    figure2.set_aspect('equal')
    
    xNormVal = norm(array(xArray))
    yNormVal = norm(array(yArray))
    for i in range(len(xArray)):
        xArray[i] = (xArray[i])/(xNormVal)
        yArray[i] = (yArray[i])/(yNormVal)
        
    figure3 = plt.subplot(133)
    figure3.plot(xArray,yArray, 'o', c= 'green')
    figure3.set_title("L2-Normalized keypoint positions")
    figure3.set_ylim([max(yArray),0])
    figure3.set_xlim([0,max(xArray)])
    figure3.set_aspect('equal')
    
    plt.rcParams["figure.figsize"] = (20,20)
    plt.show()
    return xArray,yArray, confidences
```

```python
def calculateSimilarities(x1,y1,x2,y2):
    
    #Calculate average distance between keypoints
    distances = []
    for i in range(len(x1)):
        dist = math.sqrt(abs(x1[i]-x2[i])**2+abs(y1[i]-y2[i])**2)
        distances.append(dist)
    dist = sum(distances)/len(distances)
    
    #Calculate 2*k-dim cosine similarity
    person1Vector = array(x1+y1).reshape(1,-1)
    person2Vector = array(x2+y2).reshape(1,-1)
    C = cosine_similarity(person1Vector, person2Vector)
    
    #Calculate average cosine similarity with each keypoint as its own vector
    similarities = []
    p1 =[x1, y1]
    p2 =[x2, y2]
    for i in range(len(x1)):
        c1 = array([p1[0][i],p1[1][i]]).reshape(1,-1)
        c2 = array([p2[0][i],p2[1][i]]).reshape(1,-1)
        similarities.append(cosine_similarity(c1,c2))
    Cavg = sum(similarities)/len(similarities)
    
    
    print("Average distance between keypoints: " + str(dist) + "\n\n" + str(2*len(x1))+"-dimensional cosine similarity: "+str(C[0][0])+ " or "+str(math.acos(C[0][0]))+" degrees"+"\n\nAverage cosine similarity for independent keypoint vectors: "+str(Cavg[0][0])+ " or "+str(math.acos(Cavg[0][0]))+" degrees")
```
